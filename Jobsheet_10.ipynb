Praktikum 01 - Bubble Sort

```python
def bubble_sort(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n -i - 1):
      if data[j] > data[j+1]:
        data[j], data [j+1]=data [j+1], data[j]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah diurutkan:", bubble_sort(angka))
```

Praktikum 02 - Selection Sort

```python
def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1,n):
            if data[j]<data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

Praktikum 03 - Insertion Sort

```python
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i -1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

Praktikum 06 - Sorting Barang(Nama & Harga)

```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if reverse:
                if data[j][key_index] < data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
            else:
                if data[j][key_index] > data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
    return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

3. Eksperimen Bubble Sort

```python
def bubble_sort_descending(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] < data[j + 1]:   # tanda perbandingan diganti: < (bukan >)
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

angka = [7, 5, 8]
print("Data sebelum diurutkan:", angka)
print("Data setelah diurutkan (descending):", bubble_sort_descending(angka))
```

4. Eksperimen Selection_Sort

```python
def selection_sort_largest_to_end(data):
    n = len(data)
    for i in range(n - 1, 0, -1):   # bergerak dari indeks terakhir ke awal
        max_index = 0
        for j in range(1, i + 1):
            if data[j] > data[max_index]:   # cari yang terbesar
                max_index = j
        data[i], data[max_index] = data[max_index], data[i]   # taruh terbesar ke posisi i
    return data

angka = [35, 99, 22]
print("Data sebelum diurutkan:", angka)
print("Data setelah diurutkan (descending):", selection_sort_largest_to_end(angka))
```

5.	Eksperimen Insertion Sort

```python
def insertion_sort_string(data):
    for i in range(1, len(data)):
        key = data[i]            # nama yang sedang disisipkan
        j = i - 1

        # bandingkan alfabetis: key < data[j]
        while j >= 0 and key < data[j]:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

nama = ["Nur", "Aisa", "Sean", "Annisa", "Kyla"]
print(insertion_sort_string(nama))
```

7. Analisis Kompleksitas

```python
import random
import time

#  BUBBLE SORT

def bubble_sort_count(arr):
    a = arr.copy()
    n = len(a)
    count = 0
    for i in range(n):
        for j in range(0, n - i - 1):
            count += 1   # perbandingan
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return count


#  SELECTION SORT

def selection_sort_count(arr):
    a = arr.copy()
    n = len(a)
    count = 0
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            count += 1   # perbandingan
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return count


#  INSERTION SORT

def insertion_sort_count(arr):
    a = arr.copy()
    n = len(a)
    count = 0
    for i in range(1, n):
        key = a[i]
        j = i - 1
        while j >= 0:
            count += 1  # perbandingan
            if a[j] > key:
                a[j + 1] = a[j]
                j -= 1
            else:
                break
        a[j + 1] = key
    return count


#  MAIN PROGRAM

ukuran_data = [10, 50, 100]

for ukuran in ukuran_data:
    print(f"\n=== Ukuran Input: {ukuran} ===")
    data = [random.randint(1, 1000) for _ in range(ukuran)]

    bubble = bubble_sort_count(data)
    selection = selection_sort_count(data)
    insertion = insertion_sort_count(data)

    print(f"Bubble Sort     : {bubble} perbandingan")
    print(f"Selection Sort  : {selection} perbandingan")
    print(f"Insertion Sort  : {insertion} perbandingan")
```

```python
def input_barang():
    daftar = []
    jumlah = int(input("Berapa banyak barang? "))

    for i in range(jumlah):
        print(f"\nBarang ke-{i+1}:")
        nama = input("  Nama barang : ")
        harga = float(input("  Harga barang: "))
        daftar.append({"nama": nama, "harga": harga})

    return daftar


def tampilkan_barang(data):
    print("\n=== Daftar Barang ===")
    for item in data:
        print(f"Nama: {item['nama']}, Harga: {item['harga']}")


def sort_barang(data):
    print("\nUrutkan berdasarkan:")
    print("1. Nama")
    print("2. Harga")
    pilih = input("Pilih (1/2): ")

    print("\nUrutan:")
    print("1. Ascending")
    print("2. Descending")
    arah = input("Pilih (1/2): ")

    reverse = True if arah == "2" else False

    if pilih == "1":
        return sorted(data, key=lambda x: x['nama'].lower(), reverse=reverse)
    else:
        return sorted(data, key=lambda x: x['harga'], reverse=reverse)


# PROGRAM UTAMA
barang = input_barang()

barang_sorted = sort_barang(barang)

tampilkan_barang(barang_sorted)
```
