Praktikum 01 - Sequential Search pada Data Angka

```python
def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return f"Data target ditemukan pada indeks {i}"
    return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

Praktikum 02 – Binary Search pada Data Angka

```python
def binary_search(data, target):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return f"Data {target} ditemukan pada indeks {mid}"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

Praktikum 03 – Sequential Search untuk Data Nama

```python
def sequential_search_nama(data, target):
    for i in range(len(data)):
        if data[i].lower() == target.lower():
            return f"Nama {target} ditemukan di indeks {i}"
    return f"Nama {target} tidak ditemukan"

nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data:", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Eka"))
```

Praktikum 04 – Studi Kasus Nilai Mahasiswa dengan Binary Search

Praktikum 05 – Perbandingan Waktu Sequential vs Binary Search

```python
import time

def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return -1
    return -1

def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

data = list(range(1, 100001))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search waktu:", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search waktu:", end - start, "detik")
```

3. Variasi Sequential Search

```python
def sequential_search(data, target):
    count_compare = 0  # penghitung perbandingan
    for i in range(len(data)):
        count_compare += 1  # setiap melakukan perbandingan
        if data[i] == target:
            return f"Data {target} ditemukan pada indeks {i}. Jumlah perbandingan: {count_compare}"
    return f"Data {target} tidak ditemukan. Jumlah perbandingan: {count_compare}"

angka = [4, 7, 9, 12, 15]
print("Data:", angka)
print(sequential_search(angka, 12))
```

4.	Variasi Binary Search

```python
def binary_search(data, target):
    left = 0
    right = len(data) - 1
    hitung = 0

    while left <= right:
        hitung += 1
        print("pada langkah", hitung, "belum ditemukan data", target)

        mid = (left + right) // 2

        if data[mid] == target:
            return f"data ditemukan di index {mid} dengan {hitung} langkah"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return f"data tidak ditemukan setelah {hitung} langkah"


# Program utama
angka = [50, 60, 70, 80, 90]
print("data:", angka)
print(binary_search(angka, 90))
print(binary_search(angka, 20))
```

6. Studi Kasus Barang Toko

```python
def sort_and_search_barang():
    # Data barang
    data = [
        {"nama": "Buku", "harga": 4000},
        {"nama": "Pulpen", "harga": 3000},
        {"nama": "Pensil", "harga": 2500},
    ]

    # Insertion Sort berdasarkan harga
    for i in range(1, len(data)):
        key = data[i]
        j = i - 1

        while j >= 0 and data[j]["harga"] > key["harga"]:
            data[j + 1] = data[j]
            j -= 1

        data[j + 1] = key

    # Tampilkan data terurut
    print("Daftar harga terurut (Termurah - Termahal):")
    for item in data:
        print(item)

    # Binary Search berdasarkan harga
    target = int(input("\nCari barang berdasarkan harga: "))

    start = 0
    end = len(data) - 1
    found = False

    while start <= end:
        mid = (start + end) // 2

        if data[mid]["harga"] == target:
            print("Barang ditemukan:", data[mid])
            found = True
            break
        elif data[mid]["harga"] < target:
            start = mid + 1
        else:
            end = mid - 1

    if not found:
        print("Tidak ada barang dengan harga", target)

sort_and_search_barang()
```

7. Analisis Kompleksitas

```python
import math
import time

def sequential_search_worst_case(N):
  return N
def binary_search_worst_case(N):
  if N <= 0:
    return 0
  return math.ceil(math.log2(N + 1))

ukuran_data = [100, 1000, 10000]

print("--- Analisis Kompleksitas Pencarian (Worst Case) ---")
print("{:<10} | {:<20} | {:<20}".format("Ukuran N", "Sequential Search (N)", "Binary Search (log₂ N)"))
print(" ")

tabel_pengamatan = []
for N in ukuran_data:
  seq_ops = sequential_search_worst_case(N)
  bin_ops = binary_search_worst_case(N)


  tabel_pengamatan.append({
      "N": N,
      "Sequential Ops": seq_ops,
      "Binary Ops": bin_ops
  })

  print("{:<10} | {:<20} | {:<20}".format(N, seq_ops, bin_ops))

print(" ")
```
