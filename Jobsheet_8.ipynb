Praktikum 04 - Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
    return sum(args)

def biodata(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

Praktikum 07 - Fungsi Decorator

```python
def decirator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungis  dijalankan...")
        fungsi()
        print("Setelah fungsi dijalankan...")
    return wrapper

@ decirator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")

sapa()
```

9. Fungsi Lambda

```python
import math

kuadrat = lambda x: x**2
keliling_lingkaran = lambda r: 2 * math.pi * r

bilangan = 7
jari_jari = 14
hasil_kuadrat = kuadrat(bilangan)
hasil_keliling = keliling_lingkaran(jari_jari)

print(f"Kuadrat dari {bilangan} adalah {hasil_kuadrat}")
print(f"Keliling lingkaran dengan jari-jari {jari_jari} adalah {hasil_keliling:.2f}")
```

10. Fungsi Decorator

```python
def logger(fungsi):
  def wrapper():
    print("Memulai fungsi...")
    fungsi()
    print("Fungsi selesai dijalankan...")
  return wrapper

@logger
def halo():
  print("Halo, Mahasiswa Polines")

halo()
```

11. Studi Kasus Nyata - Program Nilai

```python
def hitung_rata(data):
    return sum(data)  / len(data)

def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

input_tugas = float(input("Masukkan nilai Tugas (0-100): "))
input_ujian = float(input("Masukkan nilai Ujian (0-100): "))

nilai = [86, 90, 95]
rata = hitung_rata(nilai)
hitung_nilai_akhir = lambda nilai_tugas, nilai_ujian: (0.4 * nilai_tugas) + (0.6 * nilai_ujian)
nilai_akhir = hitung_nilai_akhir (input_tugas, input_ujian)

print("Hasil rata - rata",rata)
print("Status Kelulusan:", status_kelulusan(rata))
print(f"Nilai Akhir (40% Tugas + 60% Ujian): {nilai_akhir:.2f}")
```
